# Recuerde la regresi贸n lineal:

Como sabe, la regresi贸n lineal encuentra una funci贸n que relaciona una variable dependiente continua, y, con algunos predictores (variables independientes 1, 2, etc.). Por ejemplo, la regresi贸n lineal simple asume una funci贸n de la forma:
$$ y= \theta_0+ \theta_1 x_1 + \theta_2 x_2 + \cdots $$
y encuentra los valores de los par谩metros $ \theta_0, \theta_1, \theta_2$
 , etc., donde el t茅rmino $\theta_0$ es la "intersecci贸n". Generalmente se puede mostrar como: 
$$ h_\theta(x) = \theta^TX $$
La regresi贸n log铆stica es una variaci贸n de la regresi贸n lineal, 煤til cuando la variable dependiente observada, y, es categ贸rica. Produce una f贸rmula que predice la probabilidad de la etiqueta de clase en funci贸n de las variables independientes.

$$

# 7. matriz de confusi贸n 
Otra forma de ver la precisi贸n del clasificador es mirar la matriz de confusi贸n.

```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    Esta funci贸n imprime y traza la matriz de confusi贸n.
     La normalizaci贸n se puede aplicar configurando 'normalize=True'.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
print(confusion_matrix(y_test, yhat, labels=[1,0]))
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat, labels=[1,0])
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['TM=1','TM=0'],normalize= False,  title='Confusion matrix')
```

Mira la primera fila. La primera fila es para los egresados cuyo valor predicho de TM real en el conjunto de prueba es 1. Como puede calcular, de 427 EGRESADOS, el valor predicho de TM  de 427 de ellos es 1. Y de estos 427, el clasificador predijo correctamente 427 de ellos como 0 y 0 de ellos como 0.

Significa que nuestro clasificador no tiene error ya que para predecir la TM con un valor de 1 los clasifica todo.

El clasificador predijo correctamente 427 de ellos como 1. Lo bueno de la matriz de confusi贸n es que muestra la capacidad del modelo para predecir correctamente o separar las clases. En un caso espec铆fico de clasificador binario, como este ejemplo, podemos interpretar estos n煤meros como el recuento de verdaderos positivos, falsos positivos, verdaderos negativos y falsos negativos.

```python
print (classification_report(y_test, yhat))
```

Bas谩ndonos en el recuento de cada secci贸n, podemos calcular la precisi贸n y la recuperaci贸n (recall) de cada etiqueta:

* La precisi贸n es una medida de la precisi贸n siempre que se haya predicho una etiqueta de clase. Est谩 definido por: precision = TP / (TP + FP)

* El recall es una verdadera tasa positiva. Se define como: Recall = TP / (TP + FN)

Entonces, podemos calcular la precisi贸n y la recuperaci贸n de cada clase.

**F1 Score** : ahora estamos en condiciones de calcular los  F1 Scores para cada etiqueta en funci贸n de la precision y el recall de esa etiqueta.

La puntuaci贸n F1 es el promedio arm贸nico de la precisi贸n y la recuperaci贸n, donde una puntuaci贸n F1 alcanza su mejor valor en 1 (precisi贸n y recuperaci贸n perfecta) y el peor en 0. Es una buena forma de demostrar que un clasificador tiene un buen valor para ambos recuerdo y precisi贸n.

Y finalmente, podemos decir que la precisi贸n promedio para este clasificador es el promedio de la puntuaci贸n F1 para ambas etiquetas, que es 0,78 en nuestro caso.
