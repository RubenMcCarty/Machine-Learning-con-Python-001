Aunque la regresi贸n lineal es muy buena para resolver muchos problemas, no se puede utilizar para todos los conjuntos de datos. Primero, recuerde c贸mo la regresi贸n lineal podr铆a modelar un conjunto de datos. Modela una relaci贸n lineal entre una variable dependiente y y una variable independiente x. Ten铆a una ecuaci贸n simple, de grado 1, por ejemplo y = 2 + 3

```python
x = np.arange(-5.0, 5.0, 0.1)
# y = 2x + 3
y = 2*(x)+3
y_noise = 2* np.random.normal(size = x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot (x,y, 'r')
plt.xlabel('Varible Independiente')
plt.ylabel('Variable Dependiente')
plt.show()
```

Las regresiones no lineales son una relaci贸n entre variables independientes  y una variable dependiente  que dan como resultado datos modelados de funci贸n no lineal. B谩sicamente, cualquier relaci贸n que no sea lineal puede denominarse no lineal y, por lo general, se representa mediante el polinomio de  grados (potencia m谩xima de x).
$$ \ y = a x^3 + b x^2 + c x + d \$$

las funciones no lineales pueden tener elementos exponenciales, logaritmicas, fracciones y otros. Por ejemplo
$$ y = \log(x)$$

O incluso mas complicado como:
$$ y = \log(a x^3+ b x^2+ c x + d)$$

#### Echemos un vistaso a la grafica de funciones cubicas

# Ejemplo de regresi贸n no lineal
Por ejemplo, intentaremos ajustar un modelo no lineal a los puntos de datos correspondientes al PIB de China de 1960 a 2014. Descargamos un conjunto de datos con dos columnas, la primera, un a帽o entre 1960 y 2014, la segunda, Ingreso interno bruto anual correspondiente de China en d贸lares estadounidenses para ese a帽o.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("D:/TECHNOLOGY 2020/EIE/ML con Python/Data/china_gdp.csv")
df.head(20)
```

# Plotting the Dataset
As铆 es como se ven los puntos de datos. Parece una funci贸n log铆stica o exponencial. El crecimiento comienza lento, luego desde 2005 en adelante, el crecimiento es muy significativo. Y finalmente, se desaceler贸 levemente en la d茅cada de 2010.

```python
plt.figure (figsize=(8,5))
x_data, y_data = (df["Year"].values, df["Value"].values)
plt.plot (x_data, y_data, 'ro')
plt.ylabel("PBI")
plt.xlabel("Year")
plt.show()
```

# Elegir un modelo
A partir de una mirada inicial a la gr谩fica, determinamos que la funci贸n log铆stica podr铆a ser una buena aproximaci贸n, ya que tiene la propiedad de comenzar con un crecimiento lento, aumentar el crecimiento en el medio y luego volver a disminuir al final; como se ilustra a continuaci贸n:

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = 1.0 / (1.0 + np.exp(-X))

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

La f贸rmula de la funci贸n log铆stica es la siguiente:
$$ \hat{Y} = \frac1{1+e^{\beta_1(X-\beta_2)}}$$

$\beta_1$: Controls the curve's steepness,

$\beta_2$: Slides the curve on the x-axis.
 # Construyendo el modelo
 Construyendo el modelo de regresi贸n e inicialicemos sus par谩metros.

```python
def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y
```

Veamos una l铆nea sigmoidea de muestra que podr铆a encajar con los datos:

```python
beta_1 = 0.10
beta_2 = 1990.0

#logistic function
Y_pred = sigmoid(x_data, beta_1 , beta_2)

#plot initial prediction against datapoints
plt.plot(x_data, Y_pred*15000000000000.)
plt.plot(x_data, y_data, 'ro')
```

Nuestra tarea aqu铆 es encontrar los mejores par谩metros para nuestro modelo. Primero normalicemos nuestra xey:

```python
# Lets normalize our data
xdata =x_data/max(x_data)
ydata =y_data/max(y_data)
```

# 驴C贸mo encontramos los mejores par谩metros para nuestra l铆nea de ajuste?

podemos usar **curve_fit**, que usa m铆nimos cuadrados no lineales para ajustar nuestra funci贸n sigmoidea a los datos. Valores 贸ptimos para los par谩metros de modo que la suma de los residuos al cuadrado de sigmoide (xdata, * popt) - ydata se minimice.

Son nuestros par谩metros optimizados.

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, xdata, ydata)
#print the final parameters
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
```

Ahora trazamos nuestro modelo de regresi贸n resultante.

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

# Pr谩ctica 
驴Puedes calcular cu谩l es la precisi贸n de nuestro modelo?

```python
# write your code here

# split data into train/test
msk = np.random.rand(len(df)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

# build the model using train set
popt, pcov = curve_fit(sigmoid, train_x, train_y)

# predict using test set
y_hat = sigmoid(test_x, *popt)

# evaluation
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat - test_y) ** 2))
from sklearn.metrics import r2_score
print("R2-score: %.2f" % r2_score(y_hat , test_y) )
```
